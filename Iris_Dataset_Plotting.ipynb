**Import pandas to read a dataset, matplotlib,seaborn import for plotting and packages_version**

```
import matplotlib
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(matplotlib.__version__)
print(seaborn.__version__)
```



**To Read Iris-Dataset**

```
data=pd.read_csv('/content/DDRIVE/MyDrive/Dataset/iris-data.csv')
print(data)
```

**To find out a groupby object that contains information about the groups and the size property is used to get an int representing the number of elements in this object.**

```
data.groupby('class').size()
```

**To findout the rows and columns value in the dataset**

```
data.shape
```

**To replace the new-name with old-name**

```
data['class']=data['class'].replace(to_replace='Iris-setossa', value='Iris-setosa')
data['class']=data['class'].replace(to_replace='versicolor', value='Iris-versicolor')
```

**To drop the null value**

```
data=data.dropna()
```

```
data.groupby('class').size()
```

**To build plotting of the DataFrame**

```
data.plot()
```

**To draw data using a value through Line plotting**





```
data.plot(kind='line',subplots=True,layout=(2,4))
plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Line Plotting")
```

**To draw graphical image of the value through Box plotting**




```
data.plot(kind='box',subplots=True,layout=(2,4))
plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Box Plotting")
```

**Bar-Plotting:** 
A bar plot represents categorical data and shows comparisons among discrete categories or values.

```
data.plot(kind='bar',subplots=True,layout=(4,1))
plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Bar Plotting")
```

**Density-Plotting:**
To generate Kernel Density Estimate plot using Gaussian kernels.A Density Plot visualises the distribution of data over a continuous interval or time period.In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable.

```
data.plot(kind='density',subplots=True,layout=(2,4))
plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Density Plotting")
```

**To draw one histogram of the DataFrame’s columns.A histogram is a representation of the distribution of data. This is useful when the DataFrame’s Series are in a similar scale.**

```
data.plot(kind="hist",subplots=True,layout=(2,4))
plt.xlabel("Number")
plt.ylabel("Value")
plt.title("Histogram Plotting")
```

**To draw a scatter plot with varying marker point size and color.This kind of plot is useful to see complex correlations between two variables.**

```
plt.scatter(x=data['sepal_length_cm'],y=data['class'])
plt.title('Scatter Plotting')
```

```
x=data['class']
y=data['sepal_length_cm']
plt.subplot(1,2,1)
plt.title('subplot Plot')
plt.plot(x,y)
```

**To draw the pie chart draw for each value in the column.**

```
y = data['sepal_length_cm']
plt.pie(y)
plt.show()
```

**To draw the relationship between a numerical and one or more categorical variables using one of several visual representations onto a FacetGrid and Catplot is used for different plot kind to visualize the same data.Catplot() is used for combining a barplot() and a FacetGrid.**

```
k = sns.catplot(x="sepal_length_cm", y="sepal_width_cm", data=data, kind="bar", height=12, aspect=2, legend=False)
plt.show()
```

**To draw point estimates and confidence intervals as rectangular bars.**

```
sns.barplot(x='sepal_length_cm',y='sepal_width_cm',data=data)
```

**Subplot of barplot:**

```
plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
sns.barplot(x='sepal_length_cm',y='sepal_width_cm',data=data)
```

**The Box Plot is the visual representation of the depicting groups of numerical data through the number of data points into four parts or quarters.**

```
ax = sns.boxplot(x="sepal_length_cm", data=data)
plt.show()
```

**The lmplot plot shows the line along with datapoints on the 2d space. By specifying x and y can be set the horizontal and vertical labels repsectively.**

```
sns.lmplot(x='sepal_length_cm',y='sepal_width_cm',data=data)
```
